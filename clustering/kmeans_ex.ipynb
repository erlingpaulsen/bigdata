```python
%matplotlib inline
import numpy as np
import kmeans as km
import matplotlib.pyplot as plt
```

Creating a two dimensional dataset with four gaussian distributed clusters

```python
np.random.seed(2)
cs = 40
x1 = np.random.multivariate_normal([10,10], [[1,0],[0,1]], cs)
x2 = np.random.multivariate_normal([1,1], [[1.2,0],[0,1.3]], cs)
x3 = np.random.multivariate_normal([2,7], [[1,0],[0,1]], cs)
x4 = np.random.multivariate_normal([9,1], [[1.5,0],[0,1]], cs)

x = np.concatenate((x1,x2,x3,x4), axis=0)

fig1 = plt.figure(figsize=(12,8))
ax = fig1.add_subplot(111)
ax.scatter(x[:,0], x[:,1], marker='o', s=40)
ax.set_xlabel('$X_1$', fontsize=16)
ax.set_ylabel('$X_2$', fontsize=16)
ax.set_title('Four Gaussian distributed clusters', fontsize=20)
ax.grid(True)
```

Running k-means with $k = 4$ on the dataset.

```python
[z, means] = km.kmeans(4, x.T)
cluster = z[-1]
centroids = means[-1]

steps = np.shape(z)[0]

fig2 = plt.figure(figsize=(15,12))
fig2.suptitle('K-means clustering', fontsize=20)

ax1 = fig2.add_subplot(221)
ax1.scatter(x[:,0], x[:,1], marker='o', c=z[0], s=40)
ax1.scatter(means[0][0,:], means[0][1,:], marker='x', c='k', s=150, linewidths=2.5)
ax1.set_xlabel('$x_1$', fontsize=20)
ax1.set_ylabel('$x_2$', fontsize=20)
ax1.set_title('$1^{st}$ iteration', fontsize=16)
ax1.grid(True)

ax2 = fig2.add_subplot(222)
ax2.scatter(x[:,0], x[:,1], marker='o', c=z[1], s=40)
ax2.scatter(means[1][0,:], means[1][1,:], marker='x', c='k', s=150, linewidths=2.5)
ax2.set_xlabel('$x_1$', fontsize=20)
ax2.set_ylabel('$x_2$', fontsize=20)
ax2.set_title('$2^{nd}$ iteration', fontsize=16)
ax2.grid(True)

ax3 = fig2.add_subplot(223)
ax3.scatter(x[:,0], x[:,1], marker='o', c=z[2], s=40)
ax3.scatter(means[2][0,:], means[2][1,:], marker='x', c='k', s=150, linewidths=2.5)
ax3.set_xlabel('$x_1$', fontsize=20)
ax3.set_ylabel('$x_2$', fontsize=20)
ax3.set_title('$3^{rd}$ iteration', fontsize=16)
ax3.grid(True)

ax4 = fig2.add_subplot(224)
ax4.scatter(x[:,0], x[:,1], marker='o', c=z[3], s=40)
ax4.scatter(means[3][0,:], means[3][1,:], marker='x', c='k', s=150, linewidths=2.5)
ax4.set_xlabel('$x_1$', fontsize=20)
ax4.set_ylabel('$x_2$', fontsize=20)
ax4.set_title('Convergence', fontsize=16)
ax4.grid(True)
```
