```python
%matplotlib inline
import numpy as np
import complete_linkage as cl
import matplotlib.pyplot as plt
```

```python
x = np.array([[1,1],[1,2],[2,1],[3,5],[4,4]])

fig1 = plt.figure(figsize=(12,8))
ax = fig1.add_subplot(111)
ax.scatter(x[:,0], x[:,1], marker='o', s=80)
ax.set_xlabel('$X_1$', fontsize=16)
ax.set_ylabel('$X_2$', fontsize=16)
ax.set_title('Small example dataset', fontsize=20)
ax.grid(True)
```

```python
(z, L) = cl.complete_linkage(x)

colors1 = ['g','r','c','m','y']
colors2 = ['c','m','y','g']
colors3 = ['m','y','g']
colors4 = ['g','m']
colors5 = ['g']
fig2 = plt.figure(figsize=(15,18))
fig2.suptitle('Complete-linkage agglomerative hierarchical clustering', fontsize=20)

ax1 = fig2.add_subplot(321)
for i,cluster in enumerate(z[0]):
    for obs in cluster:
        ax1.scatter(obs[0], obs[1], marker='o', c=colors1[i], label='Cluster '+str(i+1), s=80)
ax1.set_xlabel('$x_1$', fontsize=20)
ax1.set_ylabel('$x_2$', fontsize=20)
ax1.set_title('Starting point', fontsize=16)
ax1.grid(True)
ax1.legend(loc='lower right', ncol=2, scatterpoints=1)

ax2 = fig2.add_subplot(322)
for i,cluster in enumerate(z[1]):
    for j,obs in enumerate(cluster):
        if j == 0:
            ax2.scatter(obs[0], obs[1], marker='o', c=colors2[i], label='Cluster '+str(i+1), s=80)
        else:
            ax2.scatter(obs[0], obs[1], marker='o', c=colors2[i], s=80)
ax2.set_xlabel('$x_1$', fontsize=20)
ax2.set_ylabel('$x_2$', fontsize=20)
ax2.set_title('$1^{st}$ clustering', fontsize=16)
ax2.grid(True)
ax2.legend(loc='lower right', ncol=2, scatterpoints=1)

ax3 = fig2.add_subplot(323)
for i,cluster in enumerate(z[2]):
    for j,obs in enumerate(cluster):
        if j == 0:
            ax3.scatter(obs[0], obs[1], marker='o', c=colors3[i], label='Cluster '+str(i+1), s=80)
        else:
            ax3.scatter(obs[0], obs[1], marker='o', c=colors3[i], s=80)
ax3.set_xlabel('$x_1$', fontsize=20)
ax3.set_ylabel('$x_2$', fontsize=20)
ax3.set_title('$2^{nd}$ clustering', fontsize=16)
ax3.grid(True)
ax3.legend(loc='lower right', ncol=2, scatterpoints=1)

ax4 = fig2.add_subplot(324)
for i,cluster in enumerate(z[3]):
    for j,obs in enumerate(cluster):
        if j == 0:
            ax4.scatter(obs[0], obs[1], marker='o', c=colors4[i], label='Cluster '+str(i+1), s=80)
        else:
            ax4.scatter(obs[0], obs[1], marker='o', c=colors4[i], s=80)
ax4.set_xlabel('$x_1$', fontsize=20)
ax4.set_ylabel('$x_2$', fontsize=20)
ax4.set_title('$3^{rd}$ clustering', fontsize=16)
ax4.grid(True)
ax4.legend(loc='lower right', ncol=2, scatterpoints=1)

ax5 = fig2.add_subplot(325)
for i,cluster in enumerate(z[4]):
    for j,obs in enumerate(cluster):
        if j == 0:
            ax5.scatter(obs[0], obs[1], marker='o', c=colors5[i], label='Cluster '+str(i+1), s=80)
        else:
            ax5.scatter(obs[0], obs[1], marker='o', c=colors5[i], s=80)
ax5.set_xlabel('$x_1$', fontsize=20)
ax5.set_ylabel('$x_2$', fontsize=20)
ax5.set_title('$4^{th}$ clustering', fontsize=16)
ax5.grid(True)
ax5.legend(loc='lower right', ncol=2, scatterpoints=1)
```
